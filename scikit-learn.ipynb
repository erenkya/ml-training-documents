## Intro to scikit learn

```python
import pandas as pd
import numpy as np

heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
X = heart_disease.drop("target", axis=1)
y = heart_disease["target"]
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)

clf.get_params()
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.2)
```

```python
clf.fit(X_train, y_train)
```

```python
y_label = clf.predict(np.array([0,2,3,4]))
```

```python
y_preds = clf.predict(X_test)
y_preds 
```

```python
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report , confusion_matrix , accuracy_score

print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test , y_preds)
```

```python
accuracy_score(y_test , y_preds)
```

```python
np.random.seed(42)
best_score = clf.score(X_test, y_test)
best_model = clf
for i in range (10,100,10):
      model = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
      model_score = model.score(X_test, y_test)
      if best_score < model_score:
            best_score = model_score
            best_model = model
      print(f"Random Forest Classifier with {i} estimators: {model_score}")
      print(" ")
```

### Save the model

```python
import pickle

pickle.dump(best_model , open("random_forest_model.pkl", "wb"))
```

```python
loaded_model = pickle.load(open("random_forest_model.pkl" , "rb"))
loaded_model.score(X_test, y_test)
```

## Option 2: Fill missing values with Scikit-Learn

```python
car_sales_missing = pd.read_csv("car-sales-missing-data.csv")
```

```python
car_sales_missing.dropna(subset=["Price"], inplace=True)
```

```python
X = car_sales_missing.drop("Price" , axis = 1)
y = car_sales_missing["Price"]
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

cat_imputer = SimpleImputer(strategy="constant", fill_value="missing")
num_imputer = SimpleImputer(strategy="mean")
door_imputer = SimpleImputer(strategy="constant", fill_value=4)

categorical_features = ["Make" , "Colour"]
door_feature = ["Doors"]
num_features = ["Odometer"]

imputer = ColumnTransformer([
      ("cat_imputer" , cat_imputer , categorical_features),
      ("door_imputer" , door_imputer , door_feature),
      ("num_imputer" , num_imputer , num_features)
])

filled_X = imputer.fit_transform(X)
filled_X
```

```python
car_sales_filled = pd.DataFrame(filled_X , columns=["Make", "Colour", "Doors", "Odometer"])
car_sales_filled
```
