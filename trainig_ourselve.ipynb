```python
import pandas as pd

data = pd.read_csv("mental_health_dataset.csv")
data[:5]
```

Converting String Categorical Columns IntÄ± Meaningful Number Datas

```python
from sklearn.preprocessing import LabelEncoder
categorical_cols = data.select_dtypes(include='object').columns
categorical_cols
```

```python
le = LabelEncoder()

for col in categorical_cols:
    data[col] = le.fit_transform(data[col])
    
data[:5]
```

**According to labeling High:0 Low:1 Medium:2**

```python
from sklearn.model_selection import train_test_split

X = data.drop(columns=["mental_health_risk"])
y = data["mental_health_risk"]

X_train ,X_test , y_train , y_test = train_test_split(X,y,test_size=0.2 )
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

best_n_estimator = 0
best_model = None

for estimator in range(50, 400 , 50):
    model = RandomForestClassifier(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)
    score = model.score(X_test, y_test)
    if score > best_n_estimator:
        best_n_estimator = estimator
        best_model = model
    print(f"Estimator: {estimator}\nScore: {score}\n")
print(f"Best n_estimators: {best_n_estimator}")

```

```python
y_preds = model.predict(X_test[:5])
y_preds
```
